# Object Oriented Programming - Calculation of Rental Income (individual project)

### Here we assume that we have a client coming to us asking for an automated Rental Property Calculator. Our client's name is Brandon from a company called "Bigger Pockets". Below, you will find a video of what Brandon usually does to calculate his Rental Property ROI.

Using Visual Studio Code/Jupyter Notebook, and Object Oriented Programming create a program that will calculate the Return on Investment(ROI) for a rental property.

```python
class ROI():
    """
    This class has 5 methods:
    - income: Get the rental income 
    - expenses: Calculating the rental property expenses 
    - cashFlow: Calculating the rental property's cash flow
    - cashOnCash: Calculating cash-on-cash return for the rental property 
    - getInputs: Gets all int and str inputs 
        
    """
    
    def __init__(self, rent, tax, insurance, HOA, vacancy, repair, upgrade, manage, mortgage, down_payment, closing, rehab):
        self.rent = rent
        self.tax = tax
        self.insurance = insurance
        self.HOA = HOA
        self.vacancy = vacancy
        self.repair = repair
        self.upgrade = upgrade
        self.manage = manage
        self.mortgage = mortgage
        self.down_payment = down_payment
        self.closing = closing
        self.rehab = rehab
        
        
    def income(self):
        self.rent = self.getInputs("How much do you charge for rent per month? ", int)
        
        
    def expenses(self):
        tax_input = self.getInputs("What is your property tax per month? ", int)
        self.tax = tax_input
        self.insurance = self.getInputs("How much do you pay for the insurance per month? ", int)
        self.HOA = self.getInputs("How much do you pay for HOA per month? ", int)
        vacancy_input = self.getInputs("Some poeple consider 5% of the rent for vacancy rate. Do you want to consider the vacancy rate? (y/n)? ", str)
        self.repair = self.getInputs("How much do you pay for repairing the property? ", int)
        self.upgrade = self.getInputs("How much do you pay for upgrading the property? ", int)
        self.manage = self.getInputs("How much do you pay for the property managment per month? ", int)
        self.mortgage = self.getInputs("How much is your monthly mortgage? ", int)
        if vacancy_input.lower() == 'y':
            self.vacancy = self.rent * 0.05
            self.total_exp = self.tax + self.insurance + self.HOA + self.vacancy + self.repair + self.upgrade + self.manage + self.mortgage
            return f'The total expenses for this property is ${self.total_exp}'
        elif vacancy_input.lower() == 'n':
            self.total_exp = self.tax + self.insurance + self.HOA + self.repair + self.upgrade + self.manage + self.mortgage
            return f'The total expenses for this property is ${self.total_exp}'
        
    
    def cashFlow(self):
        total_cash_flow = self.rent - self.total_exp
        return total_cash_flow
        #print(f'Your total cash flow is {total_ cash_flow}')
            
        
    def cashOnCash(self):
        #print(f'To calculate your total investment for this property, please answer the following questions: ')
        self.down_payment = self.getInputs("How much is your down payment for this property? ", int)
        self.closing = self.getInputs("how much is the closing cost on the property? ", int)
        self.rehab = self.getInputs("How much did you spend on renovating the property? ", int)
        total_inves = self.down_payment + self.closing + self.rehab
        print(f'The total investment for this property is ${total_inves}')
        annual_cash = 12 * self.cashFlow() 
        print(f'Your annual cash flow is ${annual_cash}')
        self.cash_on_cash = (annual_cash / total_inves ) * 100
        print(f'Your total cash on cash rate is {self.cash_on_cash}%')
        
    def getInputs(self, input_string, input_type):
        while True:
            try:
                return input_type(input(input_string))
            except:
                print("Please enter a valid answer")
                
                           
        
def run():
    """
    This function instantiates the class 'ROI' and asks questions from the user.
    At the end, the program tells the user if it is a great/good/bad investment.   

    """

    rental = ROI('', '', '', '', '', '', '', '', '', '', '', '')
    while True:
        print("Welcome! We help you to see if your investment on your rental property is a good invetment or a bad one!")
        user_input = rental.getInputs("Please enter 'c' and answer the following questions or if you want to exit the program, please enter 'q'", str)
        if user_input.lower() == 'q':
            break
        elif user_input.lower() == 'c':    
            rental.income()
            rental.expenses()
            rental.cashOnCash()
            if rental.cash_on_cash > 12:
                print("Congratulations! This is a great investment! ")
            elif rental.cash_on_cash >= 8 and rental.cash_on_cash <= 12:
                print("Congratulations! This is a good investment! ")
            elif rental.cash_on_cash < 8:
                print("Sorry! This is a bad investment! ")
        else:
            print("Please enter 'c' or 'q'.")
                
    print("Thanks for using our automated rental property calculator!")
        
              
    
run()        
    

```
